```python
Female_par_rapport_male=[]

for i in range(0,df.shape[0]):
    Female_par_rapport_male.append(0)

df['Vaccin_majoritaire'] = Female_par_rapport_male

for i in range (0,df.shape[0]):
 if df['CoviShield (Doses Administered)'][i]!='nan':
  if df['CoviShield (Doses Administered)'][i]>df['Covaxin (Doses Administered)'][i]:
    df['Vaccin_majoritaire'][i]=1
  if df['CoviShield (Doses Administered)'][i]<df['Covaxin (Doses Administered)'][i]:
    df['Vaccin_majoritaire'][i]=2



df.dropna(axis=0,inplace=True)
```

```python
df['State_Code'] = df['State_Code'].astype(int)
df=df.drop('CoviShield (Doses Administered)',axis=1)
df=df.drop('Covaxin (Doses Administered)',axis=1)

print(df)
x= df.iloc[:,0:4].values
y= pd.factorize(df['Vaccin_majoritaire'])[0]

print(y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
error = []
# Calculer l'erreur pour k entre 1 et 40
#Pour chaque itération, l'erreur moyenne pour les valeurs prédites
#de l'ensemble de test est calculée et sauvegardée ds la liste Erreur.
# Tester k de 1 à 40 et la distance Manhattan
for i in range(1, 25):
    knn = KNeighborsClassifier(i,metric='euclidean')
    knn_model = knn.fit(X_train, y_train)
    pred_i = knn_model.predict(X_test)
    error.append(np.mean(pred_i != y_test))
    # Append: add an object to the end of the list
   # mean: retourne= la moyenne arrithmétique des fausses prédictions (somme des erreurs sur total)
print(error)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(3)
knn_model = knn.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
u=knn.score(X_test, y_test)
print(u)

classes={0:'egalité',1:'Covishield',2:'Covix'}
x=np.array([80,80,0,3]).reshape(1,4)
pred=knn.predict(x)
print(classes[pred[0]])
```
